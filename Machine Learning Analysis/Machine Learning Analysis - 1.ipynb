# Model building

## Train the Data set <br>

Data set is trained by using sklearn 
 




```python
from sklearn.model_selection import train_test_split
```

```python
data.columns
```

```python
data.head()
```

```python
data["prediction"].value_counts()
```

```python
conversion_map = {"Iris-virginica":0, "Iris-versicolor":1,"Iris-setosa":2 }
```

```python
data["prediction"] = data["prediction"].map(conversion_map)
```

```python
data.head()
```

```python
X = data[['sepal_len', 'sepal_width', 'petal_len', 'petal_width']]
y = data['prediction']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
```

```python
X.head()
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_train.shape, X_test.shape
```

## From sklearn - DecisionTree, RandomForrest and AdaBoost Classifiers are imported for building the model of the trained dataset

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

## Build the model using DecisionTree Classifier

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

# Evaluation

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
X_test.head()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
cm =confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True)
```
