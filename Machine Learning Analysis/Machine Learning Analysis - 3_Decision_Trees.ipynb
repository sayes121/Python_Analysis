```python
import pydotplus #pip install pydotplus
from sklearn.tree import export_graphviz

def tree_graph_to_png(tree, feature_names, png_file_to_save):
    tree_str = export_graphviz(tree, feature_names=feature_names, 
                                     filled=True, out_file=None)
    graph = pydotplus.graph_from_dot_data(tree_str)  
    graph.write_png(png_file_to_save)
```

# Dendogram for redundant features

```python
from scipy.cluster import hierarchy as hc
```

```python
import scipy
```

```python
corr = np.round(scipy.stats.spearmanr(X_train).correlation, 4)
corr_condensed = hc.distance.squareform(1-corr)
z = hc.linkage(corr_condensed, method='average')
fig = plt.figure(figsize=(16,10))
dendrogram = hc.dendrogram(z, labels=X_train.columns, orientation='left', leaf_font_size=16)
plt.show()
```

# Partial dependence plots

```python
from pdpbox import pdp
from plotnine import *
```

```python
#conda install ggplot
```

```python
def plot_pdp(feat, clusters=None, feat_name=None):
    feat_name = feat_name or feat
    p = pdp.pdp_isolate(model, X_train, feat)
    return pdp.pdp_plot(p, feat_name, plot_lines=True,
                        cluster=clusters is not None,
                        n_cluster_centers=clusters)
```

```python
X_train.columns
```

```python
plot_pdp('lstat')
```

```python
plot_pdp('lstat', clusters=3)
```

```python
feats = ['lstat', 'rm']
p = pdp.pdp_interact(model, X_train, feats)
pdp.pdp_interact_plot(p, feats)
```
